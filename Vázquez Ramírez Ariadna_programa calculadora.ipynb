```python
import sympy as sp
import re

def arreglar_entrada(texto):
    # Inserta '*' entre números y letras y reemplaza '^' por '**'
    texto = re.sub(r'(\d)([a-zA-Z])', r'\1*\2', texto)
    texto = texto.replace(')(', ')*(')
    return texto.replace('^', '**').lower()

def programa_calculadora():
    x = sp.symbols('x')

    while True:
        print("\n" + " " * 60)
        print("   CALCULADORA")
        print(" " * 60)

        entrada_usuario = input("Introduce f(x): ")

        try:
            expresion = arreglar_entrada(entrada_usuario)
            f = sp.sympify(expresion)

            # Derivada y Puntos Críticos simplificados

            f1_raw = sp.diff(f, x)
            f1 = sp.factor(sp.cancel(f1_raw))
            print(f"\nf'(x) = {f1}")

            # Resolver f'(x) = 0
            puntos = sp.solve(f1, x)

            p_reales = sorted(list(set([p for p in puntos if p.is_real])))

            print(f"\nPUNTOS CRÍTICOS: {p_reales}")

            if not p_reales:
                print("No se encontraron puntos críticos reales.")
            else:
                for c in p_reales:
                    print(f"\n" + "-"*45)
                    print(f" ANALIZANDO PUNTO: x = {c}")
                    print("-"*45)

                    hallado = False
                    for n in range(2, 11):
                        # Obtenemos la derivada n-ésima limpia
                        df_n = sp.factor(sp.cancel(sp.diff(f, x, n)))
                        val = df_n.subs(x, c)

                        print(f"Orden {n}: f^({n})(x) = {df_n}")

                        res_num = float(val.evalf())
                        print(f"Evaluación: f^({n})({c}) = {res_num:.6f}")

                        if abs(res_num) > 1e-9:
                            if n % 2 == 0:
                                tipo = "MÍNIMO LOCAL" if res_num > 0 else "MÁXIMO LOCAL"
                                print(f"\n>> {tipo} en x = {c}")
                            else:
                                print(f"\n>> PUNTO DE INFLEXIÓN en x = {c}")

                            hallado = True
                            break
                        else:
                            print("COMENTARIO: Resultado 0, derivando de nuevo.")

                    if not hallado:
                        print("No se obtuvo conclusión después de 10 derivadas.")

        except Exception as e:
            print(f"\nERROR: Revisa la expresión. {e}")

        print("\n" + "="*60)
        otra = input("Quieres analizar otra función? (s/n): ").lower().strip()
        if otra != 's':
            print("Programa finalizado.")
            break

if __name__ == "__main__":
    programa_calculadora()
```
